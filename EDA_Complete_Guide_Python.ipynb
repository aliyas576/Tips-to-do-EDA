# 🏛️ PILLAR 1: Data Composition (Data ki Tarkib)

**Data Composition** ka matlab hai ke aapke data mein **kya kya cheezein** hain. Jese aap kisi biryani ko dekh ke pehchan lete hain ke isme kya kya ingredients hain -- chawal, gosht, masale, etc.

Is pillar mein hum ye dekhna seekhenge:
- 🔢 **Data Types** -- Har column ka nature kya hai? (number, text, category)
- 🕳️ **Missing Values** -- Kaunse columns mein data khaali hai?
- 🔄 **Duplicates** -- Koi naqli (duplicate) row toh nahi?
- 📋 **Basic Summary** -- Data ka Aadhaar Card -- count, mean, min, max
- 📦 **Categorical Breakdown** -- Har category mein kitne samples hain?

**🎯 Goal:** Data ke structure ko samajhna -- ke humare paas kya kya hai aur kya missing hai.

````python
# ============================================================
# 1.1 DATA TYPES -- Har Column ka Nature Samajhna
# ============================================================

print("🔍 DATA TYPES ANALYSIS:")
print("="*60)

# df.dtypes se har column ka data type pata chalta hai
# object = text/string, int64 = integer, float64 = decimal number
print(df.dtypes)

print("\n" + "-"*60)
print("📊 AUTOMATIC COLUMN CLASSIFICATION:")

# --- Categorical Columns (Text ya Categories) ---
# select_dtypes(['object', 'category']) se sirf text wale columns miltay hain
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
print(f"\n📦 Categorical Columns ({len(categorical_cols)}):")
for col in categorical_cols:
    print(f"   • {col}")

# --- Numerical Columns (Numbers) ---
# select_dtypes(['int64', 'float64']) se sirf number wale columns miltay hain
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print(f"\n🔢 Numerical Columns ({len(numerical_cols)}):")
for col in numerical_cols:
    print(f"   • {col}")

# --- Boolean Columns (True/False) ---
bool_cols = df.select_dtypes(include=['bool']).columns.tolist()
if bool_cols:
    print(f"\n✅ Boolean Columns ({len(bool_cols)}): {bool_cols}")

print("\n💡 INTERPRETATION:")
print("   → Categorical columns: One-hot encoding ya label encoding ki zaroorat")
print("   → Numerical columns: Direct use ho sakte hain models mein")
print("   → Boolean columns: Usually 0/1 mein convert kar lo")
````

````python
# ============================================================
# 1.2 MISSING VALUES -- Kaunse Columns mein Data Khaali hai?
# ============================================================

print("🕳️ MISSING VALUES ANALYSIS:")
print("="*60)

# Step 1: Missing values count karo har column ke liye
missing_count = df.isnull().sum()

# Step 2: Percentage nikalo -- kitna percent data missing hai?
missing_percentage = (missing_count / len(df) * 100).round(2)

# Step 3: Ek DataFrame banao jisme count aur percentage dono ho
missing_df = pd.DataFrame({
    'Missing_Count': missing_count,
    'Missing_Percentage': missing_percentage,
    'Data_Type': df.dtypes
})

# Sirf wo columns dikhayo jisme missing values hain (> 0)
missing_df = missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)

print(missing_df)

print("\n" + "-"*60)
print("🚨 MISSING VALUES SUMMARY:")
print(f"   • Total columns with missing values: {len(missing_df)}")
print(f"   • Most affected column: {missing_df.index[0]} ({missing_df.iloc[0]['Missing_Percentage']}% missing)")

# --- Visualization: Missing Values Heatmap ---
# Yeh heatmap batata hai ke kahan kahan data missing hai
# Yellow = missing, Dark = data present
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), 
            cbar=True,           # Color bar dikhayo
            yticklabels=False,    # Row labels hide karo (891 rows hain, clutter hoga)
            cmap='viridis',       # Color scheme
            xticklabels=True)     # Column labels dikhayo
plt.title('🕳️ Missing Values Heatmap -- Jahan yellow hai wahan data nahi!', 
          fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\n💡 INTERPRETATION:")
print("   • deck column ~77% missing hai -- shayad drop karna parega")
print("   • age ~20% missing hai -- median se fill kar sakte hain")
print("   • embarked ~0.2% missing hai -- mode se fill kar sakte hain")
````

````python
# ============================================================
# 1.3 DUPLICATES CHECK -- Koi Naqli (Duplicate) Row Toh Nahi?
# ============================================================

print("🔄 DUPLICATE ROWS ANALYSIS:")
print("="*60)

# df.duplicated() har row ke liye True/False deta hai
# .sum() se total duplicate rows ka count milta hai
duplicate_count = df.duplicated().sum()
print(f"   • Total Duplicate Rows: {duplicate_count}")

# Specific columns ke basis pe bhi check kar sakte hain
# Jese sirf 'name' aur 'age' same ho toh duplicate consider karo
key_duplicate = df.duplicated(subset=['sex', 'age', 'pclass', 'fare']).sum()
print(f"   • Key Columns Duplicates: {key_duplicate}")

if duplicate_count > 0:
    print("\n⚠️ Duplicate rows mil gayi! Hata rahe hain...")
    # drop_duplicates() se duplicates hatao, pehli row rakh lo
    df = df.drop_duplicates(keep='first')
    print(f"✅ Ab {df.shape[0]} rows reh gayi hain")
else:
    print("\n✅ Koi duplicate nahi hai, saaf suthra data hai!")
    print("   Har row unique hai -- achi baat hai!")

print("\n💡 BEST PRACTICE:")
print("   • Hamesha pehle duplicates check karo")
print("   • Agar duplicates hain toh 'first' ya 'last' rakh ke baaki hatao")
print("   • Kabhi kabhi specific columns ke basis pe duplicates check karna zaroori hota hai")
````

````python
# ============================================================
# 1.4 BASIC INFO -- Data ka Aadhaar Card
# ============================================================

print("📋 DATA INFO (Aadhaar Card):")
print("="*60)

# df.info() se pata chalta hai:
# - Kitni rows aur columns hain
# - Har column ka data type kya hai
# - Kitni non-null values hain har column mein
# - Memory usage kitni hai
df.info()

print("\n" + "="*60)
print("📊 NUMERICAL SUMMARY (df.describe()):")
print("="*60)

# df.describe() sirf numerical columns ka summary deta hai
# Count, Mean, Std (Standard Deviation), Min, 25%, 50% (Median), 75%, Max
desc = df[numerical_cols].describe().round(2)
print(desc)

print("\n" + "-"*60)
print("🎯 KEY INSIGHTS FROM SUMMARY:")
print(f"   • Average Age: {desc.loc['mean', 'age']:.1f} years")
print(f"   • Youngest Passenger: {desc.loc['min', 'age']:.1f} years")
print(f"   • Oldest Passenger: {desc.loc['max', 'age']:.1f} years")
print(f"   • Average Fare: ${desc.loc['mean', 'fare']:.2f}")
print(f"   • Most Expensive Ticket: ${desc.loc['max', 'fare']:.2f}")
print(f"   • Survival Rate: {desc.loc['mean', 'survived']:.1%}")
````

## 🧹 Data Cleaning Techniques -- Gandi Data ko Saaf Karna

**Data Cleaning** EDA ka sab se important hissa hai. Jese aap gosht saaf karte ho pehle -- usi tarah data bhi saaf karna zaroori hai!

### Common Cleaning Steps:
1. **Missing Values Handle Karna** -- Fill, Drop, ya Impute
2. **Outliers Detect Karna** -- IQR Method, Z-Score
3. **Data Type Conversion** -- String ko number mein, etc.
4. **Inconsistent Data Fix Karna** -- Spelling mistakes, case issues

**💡 Golden Rule:** Pehle missing values ka analysis karo, phir decide karo ke fill karna hai ya drop karna hai.

````python
# ============================================================
# DATA CLEANING: MISSING VALUES HANDLE KARNA
# ============================================================

print("🧹 DATA CLEANING -- MISSING VALUES:")
print("="*60)

# --- Strategy 1: Drop columns with too many missing values ---
# 'deck' column mein ~77% values missing hain -- isko drop karna behtar hai
print("Step 1: 'deck' column drop kar rahe hain (77% missing)...")
df_clean = df.drop(columns=['deck'])
print(f"   ✅ 'deck' dropped. Ab {df_clean.shape[1]} columns hain.")

# --- Strategy 2: Fill missing numerical values with median ---
# Median outliers se kam affect hoti hai compared to mean
print("\nStep 2: 'age' missing values ko median se fill kar rahe hain...")
age_median = df_clean['age'].median()
print(f"   • Age median: {age_median:.1f} years")
df_clean['age'] = df_clean['age'].fillna(age_median)
print(f"   ✅ Missing ages filled with median ({age_median:.1f})")

# --- Strategy 3: Fill missing categorical values with mode ---
# Mode = sab se zyada frequent value
print("\nStep 3: 'embarked' missing values ko mode se fill kar rahe hain...")
embarked_mode = df_clean['embarked'].mode()[0]
print(f"   • Embarked mode: '{embarked_mode}'")
df_clean['embarked'] = df_clean['embarked'].fillna(embarked_mode)
print(f"   ✅ Missing embark points filled with mode ('{embarked_mode}')")

# --- Strategy 4: Fill 'embark_town' similarly ---
town_mode = df_clean['embark_town'].mode()[0]
df_clean['embark_town'] = df_clean['embark_town'].fillna(town_mode)

# --- Verify: Ab koi missing value nahi honi chahiye ---
print("\n" + "-"*60)
print("🔍 VERIFICATION -- Koi missing value bachi?")
remaining_missing = df_clean.isnull().sum().sum()
print(f"   • Total missing values ab: {remaining_missing}")
if remaining_missing == 0:
    print("   ✅ Data bilkul saaf hai! Koi missing value nahi!")
else:
    print(f"   ⚠️ Ab bhi {remaining_missing} missing values hain")

print("\n💡 BEST PRACTICES:")
print("   • Numerical missing → Median se fill karo (outliers se safe)")
print("   • Categorical missing → Mode se fill karo")
print("   • >50% missing → Column drop karo")
print("   • <5% missing → Row drop bhi acceptable hai")
````

````python
# ============================================================
# DATA CLEANING: OUTLIERS DETECT KARNA (IQR METHOD)
# ============================================================

print("🚨 OUTLIER DETECTION -- IQR METHOD:")
print("="*60)

# IQR (Interquartile Range) Method:
# Q1 = 25th percentile, Q3 = 75th percentile
# IQR = Q3 - Q1
# Lower bound = Q1 - 1.5 * IQR
# Upper bound = Q3 + 1.5 * IQR
# Anything outside bounds = OUTLIER

def detect_outliers_iqr(data, column):
    """IQR method se outliers detect karo"""
    Q1 = data[column].quantile(0.25)  # 25th percentile
    Q3 = data[column].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound, Q1, Q3, IQR

# --- Check outliers in 'fare' ---
fare_outliers, fare_low, fare_high, fare_q1, fare_q3, fare_iqr = detect_outliers_iqr(df_clean, 'fare')

print("📊 FARE COLUMN OUTLIER ANALYSIS:")
print(f"   • Q1 (25th percentile): ${fare_q1:.2f}")
print(f"   • Q3 (75th percentile): ${fare_q3:.2f}")
print(f"   • IQR: ${fare_iqr:.2f}")
print(f"   • Lower Bound: ${fare_low:.2f}")
print(f"   • Upper Bound: ${fare_high:.2f}")
print(f"   • Outliers Found: {len(fare_outliers)} rows ({len(fare_outliers)/len(df_clean)*100:.1f}%)")

# --- Check outliers in 'age' ---
age_outliers, age_low, age_high, age_q1, age_q3, age_iqr = detect_outliers_iqr(df_clean, 'age')

print("\n📊 AGE COLUMN OUTLIER ANALYSIS:")
print(f"   • Q1 (25th percentile): {age_q1:.1f} years")
print(f"   • Q3 (75th percentile): {age_q3:.1f} years")
print(f"   • IQR: {age_iqr:.1f} years")
print(f"   • Lower Bound: {age_low:.1f} years")
print(f"   • Upper Bound: {age_high:.1f} years")
print(f"   • Outliers Found: {len(age_outliers)} rows ({len(age_outliers)/len(df_clean)*100:.1f}%)")

# --- Visualization: Box plots se outliers dikhana ---
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.boxplot(y=df_clean['fare'], ax=axes[0], color='gold')
axes[0].set_title('💰 Fare Box Plot -- Outliers dikh raha hai?', fontweight='bold')
axes[0].set_ylabel('Fare ($)')
axes[0].axhline(y=fare_high, color='red', linestyle='--', label=f'Upper Bound: ${fare_high:.2f}')
axes[0].legend()

sns.boxplot(y=df_clean['age'], ax=axes[1], color='skyblue')
axes[1].set_title('🎂 Age Box Plot -- Outliers dikh raha hai?', fontweight='bold')
axes[1].set_ylabel('Age (Years)')
axes[1].axhline(y=age_high, color='red', linestyle='--', label=f'Upper Bound: {age_high:.1f}')
axes[1].axhline(y=age_low, color='red', linestyle='--', label=f'Lower Bound: {age_low:.1f}')
axes[1].legend()

plt.suptitle('📦 Outlier Detection using IQR Method', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n💡 INTERPRETATION:")
print("   • Fare mein bohat se outliers hain -- kuch logon ne $500+ diye!")
print("   • Age mein koi extreme outlier nahi -- sab normal range mein hain")
print("   • Outliers ko handle karne ke tareeqe:")
print("     1. Cap karo (winsorization) -- max value set kardo")
print("     2. Transform karo (log transformation)")
print("     3. Drop karo (agar bahut zyada hain aur model affect kar rahe hain)")
````

# 📈 PILLAR 2: Data Distribution (Data ka Phelao)

**Data Distribution** samajhna bohat zaroori hai. Yeh batata hai ke aapke data ke values **kis tarah phaili hui** hain. Jese aap dekhte hain ke biryani mein masala har jagah equally distributed hai ya kisi jagah zyada hai.

### Is Pillar mein Seekhenge:
- 📊 **Histograms** -- Frequency distribution (kitni baar kya value aayi)
- 📦 **Box Plots** -- Outliers, quartiles, median dikhana
- 🔔 **KDE Plots** -- Smooth distribution curve
- 📐 **Skewness & Kurtosis** -- Distribution ki shape samajhna
- 🎨 **Interactive Plots** -- Plotly se hover kar ke explore karna

**🎯 Goal:** Har numerical column ka distribution samajhna -- kya normal hai, skewed hai, ya kuch aur?

````python
# ============================================================
# 2.1 AGE DISTRIBUTION -- Kitni Umar ke Log The?
# ============================================================

plt.figure(figsize=(14, 5))

# --- Left Plot: Histogram + KDE ---
plt.subplot(1, 2, 1)
sns.histplot(df_clean['age'], 
             bins=30,           # 30 bins = 30 bars
             kde=True,          # KDE curve overlay karo
             color='skyblue', 
             edgecolor='black',
             alpha=0.7)         # Thora transparent
plt.title('🎂 Age Distribution -- Histogram + KDE', fontsize=13, fontweight='bold')
plt.xlabel('Age (Years)', fontsize=11)
plt.ylabel('Count (Number of Passengers)', fontsize=11)
plt.axvline(df_clean['age'].mean(), color='red', linestyle='--', 
            label=f"Mean: {df_clean['age'].mean():.1f}")
plt.axvline(df_clean['age'].median(), color='green', linestyle='--', 
            label=f"Median: {df_clean['age'].median():.1f}")
plt.legend()

# --- Right Plot: Box Plot ---
plt.subplot(1, 2, 2)
sns.boxplot(y=df_clean['age'], color='lightcoral', width=0.3)
plt.title('📦 Age Box Plot -- Outliers dikh raha hai?', fontsize=13, fontweight='bold')
plt.ylabel('Age (Years)', fontsize=11)

plt.suptitle('👶👴 Age Analysis -- Titanic Passengers', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()

# --- Skewness Check ---
# Skewness batata hai ke distribution symmetric hai ya ek taraf jhuka hua
# skew = 0 → Perfectly symmetric
# skew > 0 → Right skewed (tail right side pe)
# skew < 0 → Left skewed (tail left side pe)
age_skew = df_clean['age'].skew()
age_kurt = df_clean['age'].kurtosis()

print("📐 DISTRIBUTION SHAPE METRICS:")
print("="*50)
print(f"   • Skewness: {age_skew:.3f}")
print(f"   • Kurtosis: {age_kurt:.3f}")
print("\n💡 INTERPRETATION:")
if age_skew > 0.5:
    print("   → Right skewed hai (buzurg zyada hain ya bachay?)")
elif age_skew < -0.5:
    print("   → Left skewed hai")
else:
    print("   → Almost symmetric hai (normal distribution jesa)")

print(f"\n📊 KEY STATS:")
print(f"   • Mean Age: {df_clean['age'].mean():.1f} years")
print(f"   • Median Age: {df_clean['age'].median():.1f} years")
print(f"   • Mode Age: {df_clean['age'].mode()[0]:.1f} years")
print(f"   • Youngest: {df_clean['age'].min():.1f} years")
print(f"   • Oldest: {df_clean['age'].max():.1f} years")
````

# 🔗 PILLAR 3: Data Relationships (Data ka Rishta)

**Data Relationships** samajhna matlab ye dekhna ke columns aapas mein **kis tarah connected** hain. Jese aap dekhte hain ke biryani mein aloo aur gosht ka rishta kya hai -- dono saath mein achay lagte hain!

### Is Pillar mein Seekhenge:
- 🔥 **Correlation Matrix** -- Numerical columns ka rishta (-1 se +1)
- 📊 **Pair Plots** -- Har column ka har column se comparison
- 🎯 **Scatter Plots** -- Do variables ka relationship visually
- 🌈 **Heatmaps** -- Visual correlation map with colors
- 🎬 **Animated Plots** -- Plotly se interactive exploration

**🎯 Goal:** Kaunse features strongly connected hain? Kaunsa feature target (survived) se sab se zyada correlated hai?

````python
# ============================================================
# 3.1 CORRELATION MATRIX -- Kaunsa Column Kis Se Dost Hai?
# ============================================================

# Correlation batata hai ke do numerical columns ka rishta kya hai
# Range: -1 (strong negative) se +1 (strong positive)
# 0 = koi rishta nahi

# Sirf numerical columns lete hain correlation ke liye
corr_cols = ['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']
corr_matrix = df_clean[corr_cols].corr()

plt.figure(figsize=(10, 8))

# Upper triangle ko mask karo -- duplicate info hide karne ke liye
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.heatmap(corr_matrix, 
            annot=True,           # Values likh do cells mein
            fmt='.2f',            # 2 decimal places
            cmap='RdYlBu_r',      # Red (negative) → Yellow (zero) → Blue (positive)
            center=0,             # Center color at 0
            square=True,         # Square cells
            linewidths=1,        # Grid lines
            cbar_kws={"shrink": .8, "label": "Correlation"},
            mask=mask,           # Upper triangle hide
            annot_kws={"size": 12, "weight": "bold"})

plt.title('🔥 Correlation Matrix -- Kaun kis se dost hai?', 
          fontsize=16, fontweight='bold', pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

print("💡 INTERPRETATION -- KEY CORRELATIONS:")
print("="*60)
print("   • survived vs fare (+0.26): Zyada fare dene wale zyada bache!")
print("   • survived vs pclass (-0.34): First class wale zyada bache!")
print("   • pclass vs fare (-0.55): Lower class = cheaper ticket (obvious!)")
print("   • age vs fare (+0.10): Umar aur fare mein thora rishta hai")
print("   • sibsp vs parch (+0.41): Jiske siblings hain unke parents bhi zyada hain")
print("\n🎯 CONCLUSION: 'pclass' aur 'fare' sab se important predictors lag rahe hain!")
````

````python
# ============================================================
# 3.2 PAIR PLOT -- Har Column ka Har Column Se Mulaqat
# ============================================================

# Pair plot har numerical column ka har numerical column se scatter plot banata hai
# Diagonal pe KDE (distribution curve) hoti hai
# Bohat useful hai multivariate relationships dekhne ke liye

# Thora sample lete hain taake plot jaldi bane (300 rows kaafi hain)
sample_df = df_clean[['survived', 'pclass', 'age', 'fare', 'sex']].dropna().sample(n=300, random_state=42)

g = sns.pairplot(sample_df, 
                 hue='sex',                    # Gender ke hisaab se color karo
                 palette={'male': '#74b9ff', 'female': '#fd79a8'},
                 diag_kind='kde',              # Diagonal pe KDE curve
                 height=2.5,                   # Har plot ki height
                 aspect=1.2,                   # Width/Height ratio
                 plot_kws={'alpha': 0.6, 's': 50, 'edgecolor': 'black'},  # Scatter plot style
                 diag_kws={'fill': True, 'alpha': 0.5})                  # KDE style

g.fig.suptitle('🔗 Pair Plot -- Har column ka har column se rishta (by Gender)', 
               fontsize=15, fontweight='bold', y=1.02)
plt.show()

print("💡 INTERPRETATION:")
print("   • Age vs Fare: No strong linear relationship")
print("   • Pclass vs Fare: Clear negative trend -- lower class = cheaper")
print("   • Diagonal KDE: Female passengers ka age distribution male se alag hai")
print("   • Survived vs Fare: Higher fare passengers zyada bache (green dots zyada upar)")
````

# ⚖️ PILLAR 4: Data Comparison (Data ka Muqabla)

**Data Comparison** ka matlab hai different groups ko **compare karna**. Jese aap compare karte hain ke Karachi ki biryani vs Lahore ki biryani -- dono mein kya farq hai?

### Is Pillar mein Seekhenge:
- 📊 **Groupby Analysis** -- Data ko groups mein baatna aur aggregate karna
- 📈 **Bar Charts** -- Categories compare karna visually
- 🎻 **Violin Plots** -- Distribution comparison across groups
- ☀️ **Sunburst & Treemap** -- Hierarchical comparison
- 🎬 **Animated Comparisons** -- Groups ko slide kar ke compare karna

**🎯 Goal:** Kaunsa group (gender, class, port) sab se zyada survive hua? Kya patterns hain?

````python
# ============================================================
# 4.1 GROUPBY ANALYSIS -- Data ko Groups Mein Baatna
# ============================================================

print("⚖️ SURVIVAL BY GENDER:")
print("="*60)

# groupby('sex') se data male/female mein divide hota hai
# .agg() se multiple operations ek saath kar sakte hain
gender_survival = df_clean.groupby('sex')['survived'].agg(
    Total='count',        # Kitne total the
    Survived='sum',       # Kitne bache (sum of 1s)
    Survival_Rate='mean'  # Mean = survival rate (0 ya 1 ka average)
).round(3)

print(gender_survival)

print("\n" + "="*60)
print("⚖️ SURVIVAL BY PASSENGER CLASS:")
print("="*60)

class_survival = df_clean.groupby('pclass')['survived'].agg(
    Total='count',
    Survived='sum',
    Survival_Rate='mean'
).round(3)

print(class_survival)

print("\n" + "="*60)
print("⚖️ SURVIVAL BY CLASS & GENDER (Cross-tabulation):")
print("="*60)

# Multiple columns se groupby -- cross-tabulation
combined = df_clean.groupby(['pclass', 'sex'])['survived'].mean().round(3)
print(combined.unstack())  # unstack() se pivot table ban jati hai

print("\n" + "="*60)
print("⚖️ SURVIVAL BY PORT OF EMBARKATION:")
print("="*60)

port_survival = df_clean.groupby('embarked')['survived'].agg(
    Total='count',
    Survived='sum',
    Survival_Rate='mean'
).round(3)
print(port_survival)

print("\n💡 INTERPRETATION:")
print("   • Female survival rate: ~74% (Bohat acha!)")
print("   • Male survival rate: ~19% (Bohat kam!)")
print("   • 1st class survival: ~63% (Ameer logon ko pehle bachaya)")
print("   • 3rd class survival: ~24% (Gareeb log peeche reh gaye)")
print("   • Cherbourg (C) se chadne wale zyada bache -- shayad zyada 1st class the?")
````

## 1️⃣ Data Cleaning Pehle (Sab se Pehle Saaf Safai)

```python
# Hamesha pehle missing values check karo
df.isnull().sum()

# Duplicates hatao
df = df.drop_duplicates()

# Outliers ko handle karo (IQR method ya Z-score)
Q1 = df['column'].quantile(0.25)
Q3 = df['column'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['column'] >= Q1 - 1.5*IQR) & (df['column'] <= Q3 + 1.5*IQR)]

# Data types check karo
df.dtypes

# Categorical ko category type mein convert karo (memory bachao)
df['category_col'] = df['category_col'].astype('category')
```

**💡 Tip:** Pehle 80% time cleaning mein lagao, baad mein analysis smooth hoga!

## 4️⃣ Ask the Right Questions (Sawal Acha Ho Toh Jawab Bhi Acha Ata Hai)

Har EDA session shuru karte waqt yeh questions poocho:

1. ❓ **Data mein kya patterns hain?** -- Trends, clusters, anomalies
2. ❓ **Koi outliers hain jo investigation deserve karte hain?** -- Data entry errors?
3. ❓ **Kaunse features target variable se strongly correlated hain?** -- Feature selection
4. ❓ **Missing data random hai ya koi pattern hai?** -- MCAR, MAR, MNAR
5. ❓ **Data balanced hai ya imbalanced?** -- Class imbalance issue?
6. ❓ **Koi multicollinearity hai?** -- Correlated independent variables?

---

## 5️⃣ Don't Overcomplicate (Simple is Beautiful)

- **Pehle simple plots se shuru karo** -- bar chart, histogram, scatter
- **Phir advanced techniques pe jao** -- pair plots, heatmaps, animations
- **Hamesha story batao** -- sirf plots dikhana kaafi nahi!
- **Color scheme consistent rakho** -- same colors for same categories
- **Labels aur titles clear rakho** -- koi guesswork nahi!

**💡 Golden Rule:** Agar aapka plot 3 second mein samajh nahi aa raha, toh wo plot theek nahi hai!

### 2️⃣ Statistical Tests
```python
from scipy import stats

# T-test: Kya male aur female survival rate mein farq hai?
male_survived = df[df['sex'] == 'male']['survived']
female_survived = df[df['sex'] == 'female']['survived']
t_stat, p_value = stats.ttest_ind(male_survived, female_survived)
print(f"P-value: {p_value}")  # < 0.05 → significant difference!

# Chi-square test: Kya class aur survival independent hain?
contingency = pd.crosstab(df['pclass'], df['survived'])
chi2, p, dof, expected = stats.chi2_contingency(contingency)
print(f"Chi-square p-value: {p}")  # < 0.05 → related!
```

### 3️⃣ Model Building

from sklearn.model_selection import train_test_split
